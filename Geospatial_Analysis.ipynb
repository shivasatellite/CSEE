Variables

Variables are used to store data that can be referenced and manipulated in program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space.  

```python
age=36
height=5.8
name="Shiva"
is_student=True

## printing the variables

print("age :", age)
print("Height:", height)
print("Name:", name)
```

```python
## Naming Conventions
## Variables names should be descriptive
## They must start with a letter or an '_' and contains letter, numbers and underscore
## variables names case sensitive

## Valid variable names
first_name = "Shiva"
last_name = "Kumar"
```

```python
# Invalid variable names
# 2age=30
#first-name="Krish"
##@name="Krish"
```

```python
## case sensitivity
name='Shiva'
Name="Kumar"
name=Name
```

```python
## Understanding variable types 
## Python is dynamically typed, type of a variable is determined at runtime

age=36  # int
height=5.8 # float
name="Shiva" # str
is_student=True  # bool

print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

```python
age=25
print(type(age))

# Type conversion
age_str=str(age)
print(age_str)
print(type(age_str))
```
