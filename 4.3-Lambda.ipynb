##### Lambda Functions in Python
Lambda functions are small anonymous functions defined using the **lambda** keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.



```python
#Syntax
lambda arguments: expression
```

```python
def addition(a,b):   ## basic function
    return a+b       ## one logic or operation you have written
```

```python
addition(2,3)
```

```python
addition=lambda a,b:a+b    #a,b - arguments  a+b - express  addition - eperation
type(addition)
print(addition(5,6))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)
```

```python
even1=lambda num:num%2==0
even1(11)
```

```python
even1=lambda num:num%2==0  ## num - parameter or argument; num place you can write a or b;   num%2==0 - expression
even1(12)
```

here i am taking multiple parameters x, y, z

```python
def addition(x,y,z):
    return x+y+z

addition(12,13,14)
```

```python
addition=lambda x,y,z:x+y+z
addition(12,13,14)
```
