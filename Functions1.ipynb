*args - Collects a variable number of positional arguments into a tuple

Useful when you don't know beforehand how many inputs will be passed.

```python
## *args is used to pass a variable number of positional arguments to a function.
def add_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(add_numbers(1, 2))
print(add_numbers(1,2,3,4,5))
```

Why we use *args
1. Flexible Functions

when you don't know how many inputs users will provide

```python
def print_names(*args):
    for name in args:
        print(name)
```

```python
print_names("Ram", "Shyam", "Sita")
print_names("A", "B", "C", "D", "E")    
```

2. Cleaner Code
Instead of writing:

```python
def func(a,b,c,d):
```

```python
# You can write:
def func(*args):
```

Difference Between *args and *kwargs
syntax           Meaning                    Type

**kwargs: Collects a variable number of keyword arguments into a dictionary

Useful when you want flexible, named parameters without fixing them in advance.

```python
def info(*args, **kwargs):
    print(args)
    print(kwargs)

info(1, 2, 3, name='Shiva', city="Mumbai")
```

Real-world Example

In Machine Learning or GIS workflows, you may write reusable functions.

```python
def process_bands(*bands):
    for band in bands:
        print(f"Processing {band}")
```

Real-world Geospatial Example

Imagine building a function to process different layers of geospatial data (roads, rivers, vegetation, buildings, etc.). Each dataset might come with operational metadata like CRS (Coordinate Reference System), resolution, or source.

```python
def process_layers(*args, **kwargs):
    print("Layers to process:", args)
    print("Metadata options:", kwargs)

# Example usage
process_layers("roads", "rivers", "vegetation", crs="EPSG:4326", resolution=30, source="Sentinel-2")
```
